### **step1)**  Read the CSV file:

```python
import pandas as pd

# read csv file from the address
Geospacial_Coordinates = pd.read_csv('http://cocl.us/Geospatial_data', sep = ',') 
# examine the shape of original input data

Geospacial_Coordinates.head()
```

### **step2)** Read the dataframe from part1 of the project and concatinate the two tables

```python
df4= pd.read_csv ('df4.csv') #import the dataframe from part1 of the project (scraped data from wikipedia)
df_final = pd.concat([df4, Geospacial_Coordinates], axis=1)
df_final = df_final .drop(['Unnamed''Postal Code'], axis = 1)
df_final .head()
```
